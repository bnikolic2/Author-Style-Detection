```python
annotator_agreement = []
for filtered_df in filtered_dataframes:
    first_5 = filtered_df.iloc[:, :5]
    row_concatenated_first_5 = [row.values for _, row in first_5.iterrows()]
    annotator_agreement.append(first_5)

annotator_agreement = pd.concat(annotator_agreement, ignore_index=True)
row_concatenated_values = [row.values for _, row in annotator_agreement.iterrows()]
annotator_f1 = []
for i, el in enumerate(row_concatenated_values):
    current_f1 = f1_score(annotator_change, row_concatenated_values[i])
    annotator_f1.append(current_f1)

avg_annotator_f1 = np.mean(annotator_f1)
print("Annotator F1 score: ", avg_annotator_f1)
```

```python
anotatori_po_verzijama = [10, 13, 9, 14, 8, 7, 7, 9]

def group_arrays(arrays, group_sizes):
    grouped = []
    start = 0
    for size in group_sizes:
        end = start + size
        grouped.append(arrays[start:end])
        start = end
    return grouped

grouped_arrays = group_arrays(row_concatenated_values, anotatori_po_verzijama)
```

```python
def create_pairwise_rating_matrix(pair, num_categories):
    num_items = len(pair[0])
    rating_matrix = np.zeros((num_items, num_categories), dtype=int)
    
    for i in range(num_items):
        rating_matrix[i, pair[0][i]] += 1
        rating_matrix[i, pair[1][i]] += 1
    
    return rating_matrix

def calculate_fleiss_kappa_for_pairs(survey, num_categories):
    pairwise_kappas = []
    pairs = list(combinations(survey, 2))
    
    for pair in pairs:
        rating_matrix = create_pairwise_rating_matrix(pair, num_categories)
        
        if np.all(rating_matrix == rating_matrix[0, :]):
            pairwise_kappas.append(np.nan)
        else:
            kappa = fleiss_kappa(rating_matrix)
            pairwise_kappas.append(kappa)
    
    pairwise_kappas = [k for k in pairwise_kappas if not np.isnan(k)]
    
    if len(pairwise_kappas) == 0:
        return np.nan
    
    avg_kappa = np.mean(pairwise_kappas)
    return avg_kappa

def calculate_average_fleiss_kappa_for_surveys(surveys):
    num_categories = 2  # Za binarne podatke (0 i 1)
    avg_kappas = []
    
    for survey in surveys:
        avg_kappa = calculate_fleiss_kappa_for_pairs(survey, num_categories)
        avg_kappas.append(avg_kappa)
    
    return avg_kappas
    

avg_kappas = calculate_average_fleiss_kappa_for_surveys(grouped_arrays)

for i, avg_kappa in enumerate(avg_kappas):
    print(f"Prosječna Fleissova kappa za anketu {i+1}: {avg_kappa}")
```

```python
df_fleiss = pd.DataFrame({
    'Survey': [f'Anketa {i+1}' for i in range(len(avg_kappas))],
    'Avg Fleiss Kappa': avg_kappas
})

df_fleiss.to_csv('avg_fleiss_kappa_results.csv', index=False)
```

```python
all_f1_scores = []

for version_idx, (predictions, ids) in enumerate(zip(all_predictions, all_ids), start=1):
    version_f1_scores = []
    print(f"Version {version_idx} Results:")
    for pred_set, id in zip(predictions, ids):
        correct_change = np.array(changes[id], dtype=int)  
        print(f"\nID: {id}")
        for pred_idx, pred in enumerate(pred_set):
            if len(pred) == len(correct_change):  
                f1 = f1_score(correct_change, pred)
                version_f1_scores.append(f1)
                print(f"    F1 Score: {f1:.4f}")
    # Average the F1 scores for this version
    if version_f1_scores:
        average_f1 = np.mean(version_f1_scores)
        print(f"\nAverage F1 Score for Version {version_idx}: {average_f1:.4f}\n")
```
